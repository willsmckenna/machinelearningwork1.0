## <h5> 1. Load and preprocess the data using Numpy or Pandas and the preprocessing functions from scikit-learn. Specifically, you need to separate the target attribute ("pep") from the portion of the data to be used for training and testing. You will need to convert the selected dataset into the Standard Spreadsheet format (scikit-learn functions generally assume that all attributes are in numeric form). Finally, you need to split the transformed data into training and test sets (using 80%-20% randomized split). [Review Ipython Notebook examples from Week 4 for different ways to perform these tasks.] </h5>

```python
target_labels = table.pep
table = table.drop(labels='pep', axis=1)
table = table.drop(labels='id', axis=1)
print(target_labels, table)
```

```python
table_dummies = pd.get_dummies(table, columns = ['gender', 'region', 'married', 'car', 'savings_acct', 'current_acct','mortgage'])
table_dummies
```

```python
from sklearn.model_selection import train_test_split
bank_train, bank_test, bank_target_train, bank_target_test = train_test_split(table_dummies, target_labels, test_size=0.2, random_state=33)

print (bank_test.shape)
bank_test[0:5]
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler().fit(bank_train)
bank_train_norm = min_max_scaler.transform(bank_train)
bank_test_norm = min_max_scaler.transform(bank_test)
```

```python
np.set_printoptions(precision=2, linewidth=80, suppress=True)
bank_train_norm[0:5]
bank_train_norm.shape
```

## <h5> 2. Run scikit-learn's KNN classifier on the test set. Note: in the case of KNN, you should first normalize the data so that all attributes are in the same scale (normalize so that the values are between 0 and 1). Generate the confusion matrix (visualize it using Matplotlib), as well as the classification report. Also, compute the average accuracy score. Experiment with different values of K and the weight parameter (i.e., with or without distance weighting) for KNN to see if you can improve accuracy (you do not need to provide the details of all of your experimentation, but provide a short discussion on what parameters worked best as well as your final results). </h5>

```python
from sklearn import neighbors, tree, naive_bayes

n_neighbors = 2

knn_classifier = neighbors.KNeighborsClassifier(n_neighbors)
knn_classifier.fit(bank_train_norm, bank_target_train)

knnpreds_test = knn_classifier.predict(bank_test_norm)
print(knnpreds_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

knn_confusion_matrix = confusion_matrix(bank_target_test, knnpreds_test)

print(knn_confusion_matrix)
print(classification_report(bank_target_test, knnpreds_test))
print (knn_classifier.score(bank_test_norm, bank_target_test))
```

## <h5> Overall, including distance weighting made accuracy not as high as without it. The best combination of parameters appears to be a K of 10 and no distance weighting with an accuracy of .69. When distance weighting is included, best accuracies tend to cluster around k=5 and without distance weighting it clusters around k=10. Without distance weighting, lower Ks appear to be less accurate then the same Ks with it. </h5>

```python
# creating tree classifier
tree_classifier = tree.DecisionTreeClassifier()
tree_classifier.fit(bank_train, bank_target_train)

treepreds_test = tree_classifier.predict(bank_test)
print(treepreds_test)

#creating Naive Bayes classifier
nb_classifier = naive_bayes.GaussianNB()
nb_classifier = nb_classifier.fit(bank_train, bank_target_train)
nbpreds_test = nb_classifier.predict(bank_test)
print (nbpreds_test)
```

```python
tree_confusion_matrix = confusion_matrix(bank_target_test, treepreds_test)

print(tree_confusion_matrix)
print(classification_report(bank_target_test, treepreds_test))

tree_accuracy_test = tree_classifier.score(bank_test, bank_target_test)
tree_accuracy_train = tree_classifier.score(bank_train, bank_target_train)

nb_confusion_matrix = confusion_matrix(bank_target_test, nbpreds_test)

print(nb_confusion_matrix)
print(classification_report(bank_target_test, nbpreds_test))

nb_accuracy_test = nb_classifier.score(bank_test, bank_target_test)
nb_accuracy_train = nb_classifier.score(bank_train, bank_target_train)

print(f"Tree test accuracy: {tree_accuracy_test}, Tree train accuracy: {tree_accuracy_train}, NB test accuracy: {nb_accuracy_test}, NB train accuracy: {nb_accuracy_train}")
```

## <h5> The above results show that overall the decision tree would be the best model to use, as it has the highest accuracy on the test set at .8, beating out KNN at .625 and Naive-bayes at .59. However, the tree has a 1.0 accuracy on the training set, showing us the model is overfit to the data and thus has high variance, and this may suggest that the model won't do as well as we would hope in future test sets. The naive-bayes model is let fit to the model, and has a high bias, we can be seen due to its relatively low accuracy on its own training set. This showed itself in the test accuracy, where it was much less accurate than the tree model. The naive-bayes might do better than expected when given a new test set compared to the tree. </h5>
